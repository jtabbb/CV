# Welcome to the AMORE file processor
The first cell contains the necessary functions to unpack the Wavedream Board binary file and produce standard plots for pulse height spectra and count rate plots.

```python
import struct
import numpy as np
import pathlib
import os, csv
from scipy import integrate as sint
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec

def parse_fheader(data):
    return data[:3], data[3:4]

def parse_theader(data):
    return data[:4]

def parse_bheader(data):
    return data[:2], struct.unpack('H', data[2:4])[0]

def parse_eheader(data):
    # I, H x 8 -> 4 + 16 = 20, but original slice uses [4:24] after reading 24
    return data[:4], struct.unpack('IHHHHHHHH', data[4:24])

def parse_tcheader(data):
    return data[:2], struct.unpack('H', data[2:4])[0]

def parse_cheader(data):
    return data[:1], data[1:4]

def quit_fn(filename,traw,successE,wfx,wfy,t,Nout,escalers_out,chi):
    print('Creating new directory...')
    filedir=os.getcwd()
    newdir=pathlib.Path(filedir,filename[:-4])
    newdir.mkdir(parents=True,exist_ok=True)
    print('Events successfully checked',successE)
    return (wfx,wfy,t,Nout,escalers_out,chi)

def rb(filename,fig1):
    """
    Same return values as rb(): (wfx, wfy, t, Nout, escalers_out, chi)
    Still writes the CSV next to the input file.
    Uses ~3-10x less RAM on large files depending on event count.
    """
    print('Reading binary file:', filename)
    wfx, wfy = [], []              # list of np.ndarray(float32, len=1024)
    t_list = []                    # list of float64 timestamps
    traw = [[], [], [], []]        # lists for hour*3600, minute*60, second, ms/1000
    escalers_out = []              # list of uint32
    chi = []                       # list of int16 channel indices
    Nout = 0
    successE = 0

    # ensure output dir
    filedir = os.getcwd()
    newdir = pathlib.Path(filedir, filename[:-4])
    newdir.mkdir(parents=True, exist_ok=True)

    # CSV path
    tempfn = filename.split('waveforms_')[1][:-4]
    out_csv_dir = os.path.join(filedir, f'waveforms_{tempfn}')
    os.makedirs(out_csv_dir, exist_ok=True)
    out_csv_path = os.path.join(out_csv_dir, f'{tempfn}_Master.csv')

    # CSV writer (streaming)
    wrote_header = False
    csv_file = open(out_csv_path, 'w', newline='')
    writer = csv.writer(csv_file)

    try:
        with open(filename, "rb") as f:
            # File header
            tag, version = parse_fheader(f.read(4))
            if tag != b'DRS' or version != b'8':
                #print(tag, version)
                csv_file.close()
                return quit_fn(filename,traw,successE,wfx,wfy,t_list,Nout,escalers_out,chi)

            # Time header
            time_header = parse_theader(f.read(4))
            if time_header != b'TIME':
                csv_file.close()
                return quit_fn(filename,traw,successE,wfx,wfy,t_list,Nout,escalers_out,chi)

            # Preallocate bin widths: unknown board count initially; grow by boards discovered
            bin_width = []  # list per board, shape (18,1024) float32

            # ---- Read board time calib (bin widths) ----
            n_boards = 0
            while True:
                bh_data = f.read(4)
                if len(bh_data) < 4:
                    break
                bn, board_serial = parse_bheader(bh_data)
                if bn != b'B#':
                    f.seek(-4, 1)
                    break

                bw_board = np.zeros((18, 1024), dtype=np.float32)
                for chn in range(18):
                    ch_data = f.read(4)
                    ch_c, ch_cn = parse_cheader(ch_data)
                    if ch_c != b'C':
                        f.seek(-4, 1)
                        break
                    #print(ch_cn)
                    ch_index = (int(chr(ch_cn[1])) * 10) + int(chr(ch_cn[2]))

                    # 1024 float32 = 4096 bytes
                    bw = np.frombuffer(f.read(1024 * 4), dtype=np.float32, count=1024)
                    bw_board[ch_index, :] = bw
                bin_width.append(bw_board)
                n_boards += 1

            # Reusable buffers
            cum_bw = np.empty(1025, dtype=np.float32)
            tm_buf = np.empty(1024, dtype=np.float32)
            wf_buf = np.empty(1024, dtype=np.float32)

            # ---- Process events ----
            while True:
                eh_data = f.read(24)
                if len(eh_data) < 24:
                    break

                event_hdr, (event_serial, year, month, day,
                            hour, minute, second, ms, v_range) = parse_eheader(eh_data)

                for b in range(n_boards):
                    bh_data = f.read(4)
                    bn, board_serial = parse_bheader(bh_data)
                    if bn != b'B#':
                        csv_file.close()
                        return quit_fn(filename,traw,successE,wfx,wfy,t_list,Nout,escalers_out,chi)

                    for chn in range(18):
                        ch_data = f.read(4)
                        ch_c, ch_cn = parse_cheader(ch_data)

                        if ch_c == b'E':  # end of board group
                            f.seek(-4, 1)
                            break
                        if len(ch_cn) < 3:
                            f.seek(-4, 1)
                            break
                        chn_index = (int(chr(ch_cn[1])) * 10) + int(chr(ch_cn[2]))

                        # scaler
                        scalers = struct.unpack('I', f.read(4))[0]
                        escalers_out.append(scalers)

                        # trigger cell
                        tc_data = f.read(4)
                        tc_tag, trigger_cell = parse_tcheader(tc_data)
                        if tc_tag != b'T#':
                            csv_file.close()
                            return quit_fn(filename,traw,successE,wfx,wfy,t_list,Nout,escalers_out,chi)

                        # voltage_raw: 1024 * uint16
                        raw = np.frombuffer(f.read(1024 * 2), dtype=np.uint16, count=1024)

                        # Build time axis with circular shift from trigger_cell
                        bw = bin_width[b][chn_index]  # (1024,) float32
                        # cum_bw[0]=0; cum_bw[1:]=cumsum(rolled)
                        cum_bw[0] = 0.0
                        # roll left by trigger_cell -> sample order starting at trigger_cell
                        rolled = np.roll(bw, -int(trigger_cell))
                        np.cumsum(rolled, out=cum_bw[1:])
                        tm_buf[:] = cum_bw[:1024]

                        # Build waveform (float32)
                        # original: voltage_raw[i] / 65536.0 - 0.9
                        wf_buf[:] = raw.astype(np.float32) / 65536.0 - 0.9

                        # Store compact copies for return
                        wfx.append(tm_buf.copy())
                        wfy.append(wf_buf.copy())

                        # Timestamp & channel bookkeeping
                        successE += 1
                        tempt = (hour * 3600) + (minute * 60) + (second) + ms / 1000.0
                        traw[0].append(hour * 3600)
                        traw[1].append(minute * 60)
                        traw[2].append(second)
                        traw[3].append(ms / 1000.0)
                        t_list.append(tempt)
                        chi.append(chn_index)
                        if chn > Nout:
                            Nout = chn

                        # ---- Feature extraction (same logic, in-place) ----
                        # Offsets
                        boff = 75
                        xoff1 = 10
                        xoff2 = 30

                        tempwf = wf_buf
                        tempwft = tm_buf

                        pkpk = float(tempwf.max() - tempwf.min())
                        # mean of some start values
                        # ref = tempwf[xoff1:xoff2].mean()  # kept but unused originally

                        # locate minimum after xoff2
                        postwf = tempwf[xoff2:]
                        acmin = xoff2 + int(np.argmin(postwf))
                        val = float(tempwf[max(acmin-3,0):acmin+3].mean())

                        hm = val + pkpk / 2.0
                        # index of value closest to half maximum after acmin
                        ind_hm = acmin + int(np.argmin(np.abs(tempwf[acmin:] - hm)))

                        base_start = tempwf[max(acmin - boff, 0):acmin]

                        if pkpk > 0.01 and base_start.size > 0:
                            try:
                                # cumulative sum trick
                                cs = np.cumsum(base_start)
                                ind_s = acmin - base_start.size + int(np.argmax(cs)) - 3
                            except ValueError:
                                boff = acmin
                                base_start = tempwf[max(acmin - boff, 0):acmin]
                                cs = np.cumsum(base_start)
                                ind_s = acmin - base_start.size + int(np.argmax(cs)) - 3

                            base_end = tempwf[-xoff2:]
                            base_mean_end = float(base_end.mean())
                            base_std_end = float(base_end.std())
                            above_thresh_end = np.where(np.abs(tempwf - base_mean_end) > 1.5 * base_std_end)[0]
                            ind_e = int(above_thresh_end[-1]) if above_thresh_end.size > 0 else (tempwf.size - 1)

                            if (ind_s < ind_hm < ind_e) and (ind_s != ind_e) and (ind_e - ind_s > 3):
                                int_offset = float(tempwf[ind_s:ind_e].max())
                                q_s = float(sint.simpson(tempwf[ind_s:ind_hm] - int_offset, x=tempwft[ind_s:ind_hm]))
                                q_t = float(sint.simpson(tempwf[ind_s:ind_e]  - int_offset, x=tempwft[ind_s:ind_e]))
                                psd = 1 - q_s / q_t if abs(q_t) >= abs(q_s) else 9999.0
                            else:
                                psd = 9999.0
                        else:
                            psd = 9999.0

                        # Stream CSV row (write header once)
                        if not wrote_header:
                            writer.writerow(['Time','Channel','PkPk','Scaler','PSD','hour','minute','second','ms'])
                            wrote_header = True
                        writer.writerow([
                            f"{tempt:.6f}",
                            int(chn_index),
                            f"{pkpk:.6f}",
                            int(scalers),
                            f"{psd:.6f}",
                            int(hour*3600),
                            int(minute*60),
                            int(second),
                            f"{ms/1000.0:.3f}",
                        ])

        print('Events successfully checked', successE)

    finally:
        try:
            csv_file.close()
        except Exception:
            pass

    # dtype-compact returns
    chi_arr = np.asarray(chi, dtype=np.int16)

    csvproc(out_csv_path,*fig1,'data1')

    return out_csv_path

vrange=np.linspace(0,1,1000)

def csvproc(csvpath,fig,axn,ax,data_label):
  figt=csvpath.split('/')[-1][:-11]
  dump=(np.array([0]),np.array([0,0]))
  data=pd.read_csv(csvpath,delimiter=',')
  time_all=data["Time"]
  chn_all=data["Channel"]
  chn_we=np.unique(chn_all)
  tzero=np.amin(time_all)
  tend=np.amax(time_all)
  trange=np.arange(0,tend-tzero,1)
  if "_FA" in csvpath or "FA" in csvpath or "Var" in csvpath or "Para" in csvpath:
    da_ind=15
    db_ind=9
    dc_ind=0
    dd_ind=13
  elif ("DM" in csvpath):# or ("PIF" in csvpath):
    da_ind=0
    db_ind=1
    dc_ind=2
    dd_ind=3

  try:
    da_pkpk=data["PkPk"][data["Channel"]==da_ind]
    da_psd=data["PSD"][data["Channel"]==da_ind]
    da_t=data["Time"][data["Channel"]==da_ind]
    da_hist=np.histogram(da_pkpk,vrange)
    da_cr=np.histogram(da_t-tzero,trange)
  except:
    print('No dE top data found')
    da_hist=dump

  try:
    db_pkpk=data["PkPk"][data["Channel"]==db_ind]
    db_psd=data["PSD"][data["Channel"]==db_ind]
    db_t=data["Time"][data["Channel"]==db_ind]
    db_hist=np.histogram(db_pkpk,vrange)
    db_cr=np.histogram(db_t-tzero,trange)
  except:
    print('No E top data found')
    db_hist=dump

  try:
    dc_pkpk=data["PkPk"][data["Channel"]==dc_ind]
    dc_psd=data["PSD"][data["Channel"]==dc_ind]
    dc_t=data["Time"][data["Channel"]==dc_ind]
    dc_hist=np.histogram(dc_pkpk,vrange)
    dc_cr=np.histogram(dc_t-tzero,trange)
  except:
    print('No E bottom data found')
    dc_hist=dump

  try:
    dd_pkpk=data["PkPk"][data["Channel"]==dd_ind]
    dd_psd=data["PSD"][data["Channel"]==dd_ind]
    dd_t=data["Time"][data["Channel"]==dd_ind]
    dd_hist=np.histogram(dd_pkpk,vrange)
    dd_cr=np.histogram(dd_t-tzero,trange)
  except:
    print('No dE bottom data found')
    dd_hist=dump



  axt=['dE Top','E Top','E Bottom','dE Bottom']
  axhist=[da_hist,db_hist,dc_hist,dd_hist]
  for j in range(4):
    axn[j].plot(axhist[j][1][:-1]*1000,axhist[j][0],drawstyle='steps-post',label=data_label)
    axn[j].set_title(axt[j])
    axn[j].set_xlabel('Voltage (mV)')
    axn[j].set_ylabel('Counts')
    axn[j].set_xlim(0,1000)
    #axn[j].set_ylim(0,max(axhist[j][0])*1.025)
    axn[j].grid()
    axn[j].legend()


  axcr=[da_cr,db_cr,dc_cr,dd_cr]
  crd=[da_cr[0],db_cr[0],dc_cr[0],dd_cr[0]]

  for j in range(4):
    ax[j].plot(axcr[j][1][:-1],axcr[j][0],marker='.',label=data_label)
    ax[j].set_title(axt[j])
    ax[j].set_xlabel('Time (s)')
    ax[j].set_ylabel('Count rate (s$^{-1}$)')
    ax[j].legend()

  return axhist,axcr

def create_fig(title):
  fig=plt.figure(figsize=(16,8),tight_layout=True)

  gs = GridSpec(4, 4, figure=fig,width_ratios=[1.5,1.5,1,1])
  ax1 = fig.add_subplot(gs[0:2, 0])
  ax2 = fig.add_subplot(gs[0:2, 1])
  ax3 = fig.add_subplot(gs[2:4, 0])
  ax4 = fig.add_subplot(gs[2:4, 1])

  ax5 = fig.add_subplot(gs[0, 2:])
  ax6 = fig.add_subplot(gs[1, 2:])
  ax7 = fig.add_subplot(gs[2, 2:])
  ax8 = fig.add_subplot(gs[3, 2:])

  axn=[ax1,ax3,ax2,ax4]
  ax=[ax5,ax6,ax7,ax8]

  fig.suptitle(title)
  return fig,axn,ax
```
